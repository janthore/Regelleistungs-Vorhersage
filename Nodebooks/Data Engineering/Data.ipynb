```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../Ergebnisse/Modellierungen/V05/Modellierung_05.csv', index_col=0, parse_dates=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

positiv = df["Deutschland (Positiv)"]
negativ = df["Deutschland (Negativ)"]
# Boxplot mit Ausreißern
plt.boxplot([positiv, negativ],labels=["Regelleistung Positiv", "Regelleistung Negativ"], patch_artist=True, showfliers=True)
plt.title("Regelleistung Boxplot")
plt.ylabel("Regelleistung in MW")
plt.grid(True)
plt.show()

```
