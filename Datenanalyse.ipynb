```python
import pandas as pd
```

```python
daten = pd.read_csv("modellierung_03.csv")
daten
```

```python
daten = daten.rename(columns={"Unnamed: 0" : "timestamp"})
daten = daten.set_index("timestamp")
```

```python
daten.index = pd.to_datetime(daten.index)
daten.index.dtype
```

```python
# Gruppieren nach Jahr
daten_by_year = daten.groupby(daten.index.year)[["Deutschland (Positiv)", "Deutschland (Negativ)"]].describe()
daten_by_year
```
